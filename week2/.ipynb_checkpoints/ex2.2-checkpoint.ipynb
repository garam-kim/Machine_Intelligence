### Part (c)
“Whiten” the data,

i.e. transform the data into 4 new uncorrelated-unit-variance variables.


Each new variable should have mean 0 and a standard deviation equal to 1. This can be
done e.g. using the transformation


V = \Sigma^(1/2) M.T X


where, 
- the new variables vi form the rows of V,


- M is a matrix containing in its columns the normalized eigenvectors of the covariance
matrix C of the centered data X \in R^(N*p) and 

- \Sigma is a diagonal matrix containing the corresponding eigenvalues

```python
data= data.T
data.shape
```

```python
c_data = data-data.mean()
```

```python
cov_data = np.cov(c_data,rowvar=True)
cov_data
```

```python
eigen_val, eigen_vec = np.linalg.eigh(cov_data)
print('eigenvalue :\n', eigen_val)
print('\neigenvector :\n', eigen_vec)
```

```python
sigma = np.diag(eigen_val**(1/2))
sigma
```

```python
# check whether the eigen vecters are normalized
np.linalg.norm(eigen_vec,axis=0)
```

```python
M = eigen_vec
```

```python
# whitening the data
w_data = (sigma@ M.T @ data).T
```

```python
w_data.head()
```

```python
fig, axs = plt.subplots(3,3,figsize=(10,10))

for i in range(3):
    for j in range(3):
        axs[i,j].scatter(w_data[i],w_data[j],color='c',alpha=0.7)
```
